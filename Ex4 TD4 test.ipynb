```python
def lire_fichier(chemin):
    with open(chemin, encoding = 'utf-8') as f:
        chaine = f.read()
    return chaine
import glob
dic_langues = {}
liste_fichiers_appr = glob.glob("corpus_multi/*/appr/*")
print("Nombre de fichiers : %i"%len(liste_fichiers_appr))
for chemin in liste_fichiers_appr:
    #print(chemin)
    dossiers = chemin.split("/")
    langue = dossiers[1]
    if langue not in dic_langues:
        dic_langues[langue] = {}
    chaine = lire_fichier(chemin)
    ngram = chaine.split()
    for n in ngram:
        if n not in dic_langues[langue]:
            dic_langues[langue][n]=1
        else:
            dic_langues[langue][n]+=1 
print(dic_langues.keys())
```
